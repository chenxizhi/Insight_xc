```python
company = pd.read_csv('crunchbase-companies.csv',encoding='latin-1')
software_company = company[company.category_code=='software']
software_company.head()
```

```python
investment = pd.read_csv('/Users/xintongchen/Downloads/crunchbase-october-2013/crunchbase-investments.csv', encoding='latin-1',low_memory=False)
rounds = pd.read_csv('/Users/xintongchen/Downloads/crunchbase-october-2013/crunchbase-rounds.csv',encoding='latin-1')
acquisitions = pd.read_csv('/Users/xintongchen/Downloads/crunchbase-october-2013/crunchbase-acquisitions.csv', encoding='latin-1')
```
